```python
def CreaTablero():
    global tablero_ini
    global nf
    global nc
    tablero_ini = []
    nf = 10
    nc = 8
    nm = 8
    # Matriz pulsado
    # rellenar matriz de 0 
    p = np.zeros((nf, nc))
    
    # Tablero inicial
    for fila in range(nf):
        tablero_ini.append([])
        for columna in range(nc):
                tablero_ini[fila].append(no_pulsado)
    
    
    
    # Matriz mina
    m = np.zeros((nf, nc))
    M2 = np.zeros((nf+2, nc+2))
    # Num minas
    a = np.zeros((nf, nc))
    # Pos de las minas
    pos = []
    for i in range(nm):
        f = random.randint(0, nf-1)
        c = random.randint(0, nc-1)
        m[f,c] = 1 
        M2[f+1,c+1] = 1 
        pos.append(tuple([f,c]))
        #pos.append(np.array(f,c))
    return(p,m,a,M2,tablero_ini)
    
ct = CreaTablero()
p  = ct[0]
m = ct[1]
a = ct[2]
M2 = ct[3]
tablero_inicial = ct[4]
```

```python

def movimiento():
    '''
    Función que pide una fila y col para hacer movimiento
    '''
    global fila
    global col 
    # Selección de fila
    while True:
        try:
            fila = int(input("Indique la fila: "))
            fila = fila-1 # Resto 1 por los índices en python
            if  fila < 0 or fila > 9:
                raise ValueError
                
            break
        except ValueError:
            print("Introduzca una fila entre 0 y 10")
    # Selección de columna
    while True:
        try:
            col =  int(input("Indique la columna: "))
            col = col-1 # Resto 1 por los índices en python
            if  col < 0 or col > 7 :
                raise ValueError
                
            break
        except ValueError:
            print("Introduzca una columna entre 0 y 8")


#movimiento() # Inicio juego
```
