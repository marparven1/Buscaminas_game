## Limpieza y preparación de los datos (Data cleaning)

### Examinación de los datos


Para la preparación de los datos debemos tener en cuenta las siguientes situaciones que podrían aparecer:
<ul>
    <li>Duplicados</li>
    <ol>
    <li>Instancias (registros duplicados) </li>
    <li>Columnas duplicadas (correlación perfecta)</li>
    </ol>
    <li>Columnas redundantes, que pueden ser obtenidas a partir de otras columnas</li>
    <li>Detectar la existencia de missing values (instancias incompletas). En este caso podemos proceder o bien eliminando esos registros (corriendo el riesgo de aparición de sesgo) o eliminando columnas completas si existe alguna con pocos datos, por lo que no aporta información a nuestro estudio, o bien imputando esos registros (puede causar problemas y error en las predicciones)
    </li>
    <li>Existencia de valores atípicos (outliers) y estudiar si es necesario eliminarlos o mantenerlos en el estudio</li>
    <li>Errores en los datos, por ejemplo: una batería negativa </li>
    <li>Plantear la creación de nuevas variables de estudio a través de las ya existentes</li>
    <li>Discretización, si fuera necesario, de variables continuas</li>
</ul>

### Valores atípicos



```python
## Cálculo del valor extremo

Q1 = data.quantile(0.25) # 1º cuartil.  No seleccionamos el ID, ni la fecha ni la variable target
Q3 = data.quantile(0.75) # Tercer cuartil
IQR = Q3 - Q1 # Recorrido intercuartílico
print('Recorrido intercuartílico: \n',IQR)
```

```python
print('Número de instancias de cada una de las variables fuera del recorrido intercuartílico: \n')
print('bat_id_node: ',(data[data['bat_id_node']<(Q1 - 1.5 * IQR)[0]].shape[0]| data[data['bat_id_node']>(Q1 + 1.5 * IQR)[0]].shape[0] ))
print('bat_level: ',data[data['bat_level']<(Q1 - 1.5 * IQR)[1]].shape[0] | data[data['bat_level']>(Q1 + 1.5 * IQR)[1]].shape[0] )
```

```python
Importe = data['bat_level']
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
p1=sns.boxplot(x=Importe,palette='Paired',showfliers = True)
p1.set_title("Nivel de la batería con outliers")
plt.subplot(1,2,2)
p2=sns.boxplot(x=Importe,palette='Paired',showfliers = False)
p2.set_title("Nivel de la batería sin outliers")

plt.show()
```

Todos los outliers corresponden a un nivel de batería superior al 40%.

#### Variables categóricas

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)

g = sns.countplot(data =data, x = 'nod_name',
              order = data.nod_name.value_counts().index,
              palette="vlag")
g.set(title='Número de registros según nodo',xlabel='Nombre del nodo',ylabel='Número de casos')
for container in g.containers:
    tmp_hue = data.loc[data['nod_name']==container.get_label()]
    g.bar_label(container, labels=tmp_hue['nod_name'])


plt.subplot(1,2,2)
g2 = sns.countplot(data =data, x = 'par_name',
              order = data.par_name.value_counts().index,
              palette="vlag")
g2.set(title='Número de registros según parcela',
      xlabel='Nombre del nodo',
      ylabel='Número de casos')
for container in g2.containers:
    tmp_hue = data.loc[data['par_name']==container.get_label()]
    g2.bar_label(container, labels=tmp_hue['par_name'])
plt.show()
```

Número de registors de lopy2 bastante inferior al resto.

La mayoría de nodos son de tipo control, el doble.

```python
plt.figure(figsize=(20,5))

g = sns.countplot(data =data, x = 'nod_name', hue = 'par_name',
              order = data.nod_name.value_counts().index,
              palette="vlag")
g.set(title='Número de registros de cada nodo',
      xlabel='Nombre del nodo',
      ylabel='Número de casos')
for container in g.containers:
    tmp_hue = data.loc[data['nod_name']==container.get_label()]
    g.bar_label(container, labels=tmp_hue['nod_name'])

```

```python
plt.figure(figsize=(20,5))
fg = sns.catplot('far_name', data=data, kind='count', aspect=1.5,
                order=data.far_name.value_counts().index,
                 palette="vlag")

fg.set_xlabels('Nombre de la granja')
fg.set_ylabels('Número de registros')
fg.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
fg.fig.suptitle('')
plt.show()
```
